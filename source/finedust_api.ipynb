```python
from bs4 import BeautifulSoup   
import pandas   
soup = BeautifulSoup(response.text, 'html.parser') #line1
ItemList=soup.findAll('item') #line2
```

```python
import pandas

datetimeList=[]
stationnameList=[]
khaigradeList=[]
pm10valueList=[]
pm10gradeList=[]
pm25valueList=[]
pm25gradeList=[]


for item in ItemList:
    datetime=item.find('datatime').text 
    stationname=item.find('stationname').text
    khaigrade=item.find('khaigrade').text
    pm10value=item.find('pm10value').text
    pm10grade=item.find('pm10grade').text
    pm25value=item.find('pm25value').text
    pm25grade=item.find('pm25grade').text
    
    datetimeList.append(datetime) 
    stationnameList.append(stationname)
    khaigradeList.append(khaigrade)
    pm10valueList.append(pm10value)
    pm10gradeList.append(pm10grade)
    pm25valueList.append(pm25value)
    pm25gradeList.append(pm25grade)
```

```python
DATA = pandas.concat([pandas.DataFrame(datetimeList),
pandas.DataFrame(stationnameList),
pandas.DataFrame(khaigradeList),
pandas.DataFrame(pm10valueList),                      
pandas.DataFrame(pm10gradeList),
pandas.DataFrame(pm25valueList),
pandas.DataFrame(pm25gradeList)]
,axis=1) 
```
